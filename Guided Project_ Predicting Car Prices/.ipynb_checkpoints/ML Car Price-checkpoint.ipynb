## Predicting Car Prices 

We received the dataset **imports-85.data**. It contains information about vatious cars like weight, miles, per gallon and etc. This workbook will dedicate the fundamentals of k-nearest neighbors algorithm.

```python
#import

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

This dataframe don't have a columns. Then we can declare them.

```python
cols = ['symboling','normalized_losses','make','fuel_type','aspiration','num_doors','body_style','drive_wheels','engine_location', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_type', 
        'num_of_cylinders', 'engine_size', 'fuel_system', 'bore', 'stroke', 'compression_rate', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']
```

```python
#open the dataset

cars = pd.read_csv('imports-85.data', names=cols)
cars.head()
```

Lets choose only the columns with continuos values.

```python
numeric_cars = cars[['normalized_losses', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_size', 'bore', 'stroke', 'compression_rate', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']]
numeric_cars
```

### Data Cleaning

```python
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars.head()
```

```python
#transform types to float

numeric_cars = numeric_cars.astype('float')
numeric_cars.head()
```

```python
numeric_cars.isnull().sum()
```

For next steps of observations we need to choose method to working with missing values. If we will keep them, our result may be wrong.

We can drop the columns with missing values, drop the rows with missing values and fill the missing values using average meanong or something similar.

For this project we delete rows with null price values, because price is our meaninw we want to predict. Other columns we fill the meaning values.

```python
#delete the rows with null price and fill the cells with average values

numeric_cars = numeric_cars.dropna(subset=['price'])
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
```

```python
#check null values

numeric_cars.isnull().sum()
```

```python
numeric_cars.head()
```

Afer that we must to normalize all values.

```python
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min()) / (numeric_cars.max()- numeric_cars.min())
numeric_cars['price'] = price_col
numeric_cars.head()
```

### Univariate K-nearest neighbors models

We will start with simple model before moving to complex to understand the features better.

```python
#create the function 

def knn_train_test(train, target, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)
    
    snuffled_index = np.random.permutation(df.index)
    rand_df=df.reindex(snuffled_index)
    
    half = int(len(rand_df)/2)
    train_df = rand_df.iloc[0:half]
    test_df = rand_df.iloc[half:]
    
    knn.fit(train_df[[train]], train_df[[target]])
    
    prediction = knn.predict(test_df[[train]])
    mse = mean_squared_error(test_df[[target]], prediction)
    rmse = mse**(1/2)
    return round(rmse, 3)
```

```python
#apply the functions for different parameters

cols_to_ml = numeric_cars.columns[:-1]
cols_to_ml
```

```python
pred_results = {}

for i in cols_to_ml:
    pred_results[i] = knn_train_test(i, 'price', numeric_cars)
    
pred_results
```

```python
rmse_results_series = pd.Series(pred_results)
rmse_results_series.sort_values()
```

As the results we have a different RMSEs from models with unique charactersric with defaut k-value. The best result falls on ***engine-size***.

Next we refresh the function for different k-values.

```python
#modify the function 

def knn_train_test(train, target, df):
    np.random.seed(1)
    snuffled_index = np.random.permutation(df.index)
    rand_df=df.reindex(snuffled_index)
    
    half = int(len(rand_df)/2)
    train_df = rand_df.iloc[0:half]
    test_df = rand_df.iloc[half:]
    
    k = [i for i in range(1,10,2)]
    k_rmses = {}
    for i in k:
        knn = KNeighborsRegressor(n_neighbors = i)
        knn.fit(train_df[[train]], train_df[target])
        prediction = knn.predict(test_df[[train]])
        mse = mean_squared_error(test_df[target], prediction)
        rmse = mse**(1/2)
        k_rmses[i] = rmse

    return k_rmses
```

```python
pred_results_k = {}

for d in cols_to_ml:
    pred_results_k[d] = knn_train_test(d, 'price', numeric_cars)
    
pred_results_k
```

Lets make a line plot for visualizing results. 

```python
plt.style.use('fivethirtyeight')
for a,b in pred_results_k.items():
    x = list(b.keys())
    y = list(b.values())
    
    plt.plot(x,y)
    plt.xlabel("K")
    plt.ylabel('RMSE')
    plt.text(y=10500, x = 0, s='Root Mean Squared Error for Univariate Model with K = 1-9', color='red')
```

### K-Fold Validation

Now let's apply another validation method. KFold divides all the samples in k groups of samples, called folds. One fold is test data and other is training data.

```python
#import modules
from sklearn.model_selection import cross_val_score, KFold
```

As with the previous model we should modify the functions. For this validation we take constant k value as 5 and number of folds equal also 5.

```python
def knn_train_test(train, target, df):
    kf = KFold(5, shuffle=True, random_state=1)
    knn = KNeighborsRegressor(n_neighbors=5)
    mses = cross_val_score(knn,df[train], df[target], scoring='neg_mean_squared_error',cv=kf)
    rmses = np.sqrt(np.absolute(mses))
    avg_rmse = np.mean(rmses)
    return avg_rmse
```

```python
#find rmses for best_features - engine_size and horsepower

best_features_rmses_kfold = knn_train_test(best_features, 'price', numeric_cars)
best_features_rmses_kfold
```

The result is near with previous, but we repeat all the previous step with hyperparameters and different combination of features. For this we modify function again.

```python
def knn_train_test(train, target, df):
    kf = KFold(5, shuffle=True, random_state=1)
    
    k_rmses = {}
    
    for i in range(1,10):
        knn = KNeighborsRegressor(n_neighbors = i)
        mses = cross_val_score(knn,df[train], df[target], scoring='neg_mean_squared_error',cv=kf)
        rmses = np.sqrt(np.absolute(mses))
        avg_rmse = np.mean(rmses)
        k_rmses[i] = avg_rmse
        
    return k_rmses
```

```python
mult_rmse_kfold = {}
b = []

for i in range(0, len(features_for_mult)):
    b.append(features_for_mult[i])
    print(b)
    if i == 0:
        continue
    mult_rmse_kfold[i+1] = knn_train_test(b, 'price', numeric_cars)
```
